```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

Unbalanced data

0 --> Normal transaction

1 --> fraudulent

```python
genuine = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
genuine.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_data.groupby('Class').mean()
```

Sample Data

```python
genuine_sample = genuine.sample(n = 492)
```

new Data Frame

```python
new_data = pd.concat([genuine_sample, fraud], axis = 0)
```

```python
new_data.head()
```

```python
new_data.shape
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X = new_data.drop(columns = 'Class', axis = 1)
Y =new_data['Class']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
training_data_f1Score = f1_score(X_train_prediction, Y_train)
training_data_recall_score = recall_score(X_train_prediction, Y_train)
training_data_precision_score = precision_score(X_train_prediction, Y_train)



print("Accuracy for training data:", training_data_accuracy)
print("F1 score for training data:",training_data_f1Score )
print("recall score for training data:",training_data_recall_score  )
print("precision score for training data:",training_data_precision_score )
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
testing_data_f1Score = f1_score(X_test_prediction, Y_test)
testing_data_recall_score = recall_score(X_test_prediction, Y_test)
testing_data_precision_score = precision_score(X_test_prediction, Y_test)


print("Accuracy for training data:", testing_data_accuracy)
print("F1 score for testing data:",testing_data_f1Score )
print("recall score for testing data:",testing_data_recall_score )
print("precision score for testing data:",testing_data_precision_score )
```

Another model

```python
classifier = RandomForestClassifier()
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print("Accuracy for training data:", training_data_accuracy)

```

```python
X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print("Accuracy for training data:", testing_data_accuracy)
```
